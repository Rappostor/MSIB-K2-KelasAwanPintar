# Metode Elbow

Sebelum men-training model K-Means, sebaiknya kita mencari nilai K yang paling baik/optimal terlebih dahulu.

Nilai ini bisa kita cari menggunakan Metode Elbow atau `KElbowVisualizer` pada Python.



image.png

```python
# Tulis di sini
tes_model = KMeans(random_state=42)
visualizer = KElbowVisualizer(tes_model, k=(2,10))
visualizer.fit(X)
visualizer.show()
```

Berdasarkan hasil Metode Elbow, Titik Elbow berada pada K=5.

Maka nilai K terbaik/optimal adalah 5.

# Modeling

Selanjutnya kita lakukan training model menggunakan algoritma K-Means dengan:
- cluster bejumlah 5 (K=5)
- random state bernilai 42

Random state berguna untuk mengontrol ke-random-an inisiasi centroid.


Kita juga bisa men-tuning hyperparameter K-Means lainnya. Baca dokumentasi lengkapnya [di sini](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html).

image.png

```python
# Tulis di sini
jumlah_cluster = 5
randomizer = 42
model_kmeans = KMeans(n_clusters=jumlah_cluster, random_state=randomizer)

model_kmeans.fit(X)
```

image.png

```python
# Tulis di sini
labels_kmeans = model_kmeans.labels_
labels_kmeans
```

Karena kita menentukan K=5 (cluster berjumlah 5), maka kita akan mendapat 5 centroid.

Informasi (letak) tiap centroid bisa kita akses menggunakan method `cluster_centers_`.

image.png

```python
# Tulis di sini
print(model_kmeans.cluster_centers_)
```

## Silhouette Coefficient

Kita bisa menggunakan `silhouette_score` untuk mendapat Silhouette Score model yang telah kita training.

image.png

```python
# Tulis si sini
SC_kmeans = silhouette_score(X, labels_kmeans, metric='euclidean')

print('Silhouette Score model K-means= ', SC_kmeans)
```

Semakin Silhouette Score mendekati 1, maka model clustering akan semakin bagus.

Berdasarkan Silhouette Score, model K-Means yang telah kita training sudah cukup bagus.

Alangkah baiknya jika kita men-training lebih dari satu model clustering dengan berbagai algoritma dan tuning hyperparameter.

Kemudian, kita bandingkan Silhouette Score tiap model untuk mengetahui mana model terbagus.

## Activity 1

- Gunakan dataset Mall_Customers.csv
- Gunakan variable Age, AnnualIncome, dan SpendingScore
- Lakukan clustering menggunakan model AHC dan K-Means
- Tentukan nilai K terbaik untuk setiap model
- Bandingkan SC model K-Means dengan AHC
- Tarik kesimpulan activity dan paparkan mana model terbaik

```python
# Import library yang diperlukan
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score

# 1. Memuat Dataset
data = pd.read_csv('Mall_Customers.csv')

# 2. Memilih fitur Age, AnnualIncome, dan SpendingScore
X = data[['Age', 'AnnualIncome', 'SpendingScore']]

# 3. Standarisasi fitur untuk memastikan setiap fitur berkontribusi secara proporsional
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 4. Clustering menggunakan K-Means
# Rentang nilai K yang akan dicoba
nilai_k = range(2, 6)

# Variabel untuk menyimpan K terbaik dan Silhouette Score yang sesuai
k_terbaik = 0
silhouette_score_terbaik_kmeans = -1
label_kmeans_terbaik = None

# Mencari K terbaik untuk K-Means menggunakan Silhouette Coefficient
for k in nilai_k:
    kmeans = KMeans(n_clusters=k, random_state=42)
    label_kmeans = kmeans.fit_predict(X_scaled)
    silhouette_avg_kmeans = silhouette_score(X_scaled, label_kmeans)
    print(f"K-Means -> K={k}, Silhouette Score={silhouette_avg_kmeans}")

    if silhouette_avg_kmeans > silhouette_score_terbaik_kmeans:
        silhouette_score_terbaik_kmeans = silhouette_avg_kmeans
        k_terbaik = k
        label_kmeans_terbaik = label_kmeans

print(f"\nK terbaik untuk K-Means: {k_terbaik}, dengan Silhouette Score: {silhouette_score_terbaik_kmeans}\n")

# 5. Clustering menggunakan Agglomerative Hierarchical Clustering (AHC)
# Variabel untuk menyimpan hasil terbaik dari AHC
silhouette_score_terbaik_ahc = -1
label_ahc_terbaik = None
cluster_ahc_terbaik = 0

# Mencari jumlah cluster terbaik untuk AHC menggunakan Silhouette Coefficient
for k in nilai_k:
    agglo_clustering = AgglomerativeClustering(n_clusters=k, metric='euclidean', linkage='ward')
    label_agglo = agglo_clustering.fit_predict(X_scaled)
    silhouette_avg_ahc = silhouette_score(X_scaled, label_agglo)
    print(f"AHC -> K={k}, Silhouette Score={silhouette_avg_ahc}")

    if silhouette_avg_ahc > silhouette_score_terbaik_ahc:
        silhouette_score_terbaik_ahc = silhouette_avg_ahc
        label_ahc_terbaik = label_agglo
        cluster_ahc_terbaik = k

print(f"\nK terbaik untuk AHC: {cluster_ahc_terbaik}, dengan Silhouette Score: {silhouette_score_terbaik_ahc}\n")

# 6. Visualisasi Hasil Clustering Terbaik untuk K-Means dan AHC
plt.figure(figsize=(12, 5))

# Visualisasi K-Means
plt.subplot(1, 2, 1)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=label_kmeans_terbaik, cmap='viridis')
plt.title(f"K-Means Clustering (K={k_terbaik})")
plt.xlabel('Umur (distandardisasi)')
plt.ylabel('Pendapatan Tahunan (distandardisasi)')

# Visualisasi AHC
plt.subplot(1, 2, 2)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=label_ahc_terbaik, cmap='viridis')
plt.title(f"Clustering AHC (K={cluster_ahc_terbaik})")
plt.xlabel('Umur (distandardisasi)')
plt.ylabel('Pendapatan Tahunan (distandardisasi)')

plt.tight_layout()
plt.show()

# 7. Kesimpulan dan Perbandingan Silhouette Coefficient
print(f"Silhouette Score terbaik untuk K-Means: {silhouette_score_terbaik_kmeans}")
print(f"Silhouette Score terbaik untuk AHC: {silhouette_score_terbaik_ahc}")

# Tentukan model terbaik berdasarkan Silhouette Score
if silhouette_score_terbaik_kmeans > silhouette_score_terbaik_ahc:
    print("\nK-Means adalah model terbaik berdasarkan Silhouette Score.")
else:
    print("\nAHC adalah model terbaik berdasarkan Silhouette Score.")
```

## Activity 2

1. Buat kelompok
2. Carilah dataset untuk kasus clustering di sekitarmu atau di internet
3. Lakukan clustering menggunakan algortima K-Means (ikuti step pada hands-on ini)
4. Buat kesimpulan clustering
5. Kumpulkan ke HC

```python
# Import library yang diperlukan
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# 1. Memuat Dataset
data = pd.read_csv('Live.csv')

# 2. Memilih fitur numerik yang relevan
X = data[['num_reactions', 'num_comments', 'num_shares', 'num_likes', 'num_loves', 'num_wows', 'num_hahas', 'num_sads', 'num_angrys']]

# 3. Standarisasi fitur untuk skala yang proporsional
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 4. Clustering menggunakan K-Means
# Rentang nilai K yang akan dicoba
nilai_k = range(2, 6)

# Variabel untuk menyimpan K terbaik dan Silhouette Score yang sesuai
k_terbaik = 0
skor_silhouette_terbaik = -1
label_kmeans_terbaik = None

# Mencari K terbaik untuk K-Means menggunakan Silhouette Coefficient
for k in nilai_k:
    kmeans = KMeans(n_clusters=k, random_state=42)
    label = kmeans.fit_predict(X_scaled)
    rata_silhouette = silhouette_score(X_scaled, label)
    print(f"K={k}, Silhouette Score={rata_silhouette}")

    if rata_silhouette > skor_silhouette_terbaik:
        skor_silhouette_terbaik = rata_silhouette
        k_terbaik = k
        label_kmeans_terbaik = label

print(f"\nK terbaik untuk K-Means: {k_terbaik}, dengan Silhouette Score: {skor_silhouette_terbaik}")

# 5. Visualisasi Hasil Clustering
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=label_kmeans_terbaik, cmap='viridis', s=50)
plt.title(f"K-Means Clustering (K={k_terbaik})")
plt.xlabel('num_reactions (scaled)')
plt.ylabel('num_comments (scaled)')
plt.show()

# 6. Menambahkan hasil clustering ke dalam dataset
data['Cluster'] = label_kmeans_terbaik

# 7. Menampilkan deskripsi rata-rata setiap cluster
for i in range(k_terbaik):
    print(f"\nCluster {i}:")
    print(data[data['Cluster'] == i][['num_reactions', 'num_comments', 'num_shares', 'num_likes', 'num_loves', 'num_wows', 'num_hahas', 'num_sads', 'num_angrys']].mean())

# 8. Kesimpulan Clustering
print("\nKesimpulan Clustering:")
print("Clustering ini membagi postingan media sosial berdasarkan reaksi seperti likes, loves, wows, hahas, dll.")
print("Setiap cluster merepresentasikan pola interaksi dan reaksi yang berbeda dari pengguna terhadap postingan.")
```
